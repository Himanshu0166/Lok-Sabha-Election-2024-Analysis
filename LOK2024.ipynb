# <font color="orange"><span style="font-family:Courier;">**📌 Election Data Analysis** </font></span>

This program loads and analyzes election-related datasets using **Pandas**, **Matplotlib**, and **Seaborn**.  
It imports data from three Excel files:  

- `candidates_with_phase.xlsx`: Contains information about candidates and election phases.  
- `result.xlsx`: Includes election results for all candidates.  
- `result_winner.xlsx`: Lists the winners of the election.  

The program reads these datasets into Pandas DataFrames and displays the first few rows of each for initial inspection.  



The dataset used in this analysis is sourced from Kaggle. You can download it from the following link:  

[Download the dataset from Kaggle](https://www.kaggle.com/datasets/jainaru/lok-sabha-2024-results-and-candidates?resource=download)



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Excel files
file1 = "candidates_with_phase.csv"
file2 = "results_2024.csv"
file3 = "results_2024_winners.csv"

df_candidates = pd.read_csv(file1)
df_results = pd.read_csv(file2)
df_winners = pd.read_csv(file3)

# Show first few rows of each dataset
print("\033[91mCandidates Dataset:\n\033[0m", df_candidates.head())
print("\033[91m\nResults Dataset:\n\033[0m", df_results.head())
print("\033[91m\nWinners Dataset:\n\033[0m", df_winners.head())
```

## <font color="darkblue"><span style="font-family:Courier;">**📌 Gender Distribution of Candidates**</font></span>

This visualization displays the gender distribution of candidates in the Lok Sabha 2024 elections.  
The dataset is analyzed using **Pandas**, and the count plot is created using **Seaborn** with customized aesthetics.  

### <font color="darkblue"><span style="font-family:Courier;">**Key Features of the Visualization:**</font></span>
- **Seaborn's whitegrid style** is used for a clean and structured appearance.  
- **Count plot** represents the number of male and female candidates.  
- **Color customization** using the 'viridis' palette for better contrast.  
- **Edge highlighting** and **data labels** on bars to enhance readability.  
- **Custom title, axis labels, and grid styling** for better visualization.  

This chart helps in understanding the gender representation among the candidates. 


```python
# Set Seaborn style
sns.set_style("whitegrid")

# Create the figure
plt.figure(figsize=(7, 5),facecolor='lightgray')  # Slightly bigger for better visibility

# Customizing the count plot
ax = sns.countplot(
    data=df_candidates,
    x="Gender",
    hue='Gender', palette='viridis', legend=False,
    edgecolor="black",  # Adds black border to bars
    linewidth=1.5  # Thickens the bar edges
)
ax.set_facecolor('lightgreen')
# Add data labels (annotations)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Customize title and labels
plt.title(" Gender Distribution of Candidates ", fontsize=14, fontweight='bold', color="darkblue")
plt.xlabel("Gender", fontsize=12, fontweight='bold', color="darkgreen")
plt.ylabel("Count", fontsize=12, fontweight='bold', color="darkred")

# Customize gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

## <font color="darkblue"><span style="font-family:Courier;">**📌 Age Distribution of Candidates**</font></span>  

This visualization represents the age distribution of candidates in the **Lok Sabha 2024 elections**.  
The histogram provides insights into the spread and concentration of candidate ages, helping identify common age groups.  

###  <font color="darkblue"><span style="font-family:Courier;">**Key Features of the Visualization:**</font></span>  
- **Histogram with KDE (Kernel Density Estimation)**: Displays the age distribution along with a smooth density curve for better analysis.  
- **Seaborn's darkgrid style**: Enhances readability with a subtle background grid.  
- **Customized aesthetics**:  
  - **Color:** Blue histogram (`#1f77b4`) with a purple KDE curve for contrast.  
  - **Transparency (alpha=0.8)**: Ensures a sleek look.  
  - **Black edge color** for bars with increased linewidth for clarity.  
- **Annotations on bars**: Displays the count of candidates in each bin.  
- **Title and labels**: Styled with custom font size, weight, and colors.  
- **Gridlines and ticks**: Improve readability while maintaining a clean layout.  

This plot helps in understanding the age demographics of the candidates, showing which age groups have higher representation in the elections.  


```python
# Set Seaborn style
sns.set_style("darkgrid")

# Create the figure
plt.figure(figsize=(9, 5),facecolor='lightgray')  # Increased size for better readability

# Customize histogram
ax = sns.histplot(
    df_candidates["Age"],
    bins=15,  # Slightly reduced bins for clarity
    kde=True,  # Include Kernel Density Estimation (smooth curve)
    color="#1f77b4",  # Nice shade of blue
    alpha=0.8,  # Transparency for a sleek look
    edgecolor="black",  # Outline for each bar
    linewidth=1.2
)

ax=sns.kdeplot(df_candidates["Age"], color="purple", linewidth=2, linestyle="--")
ax.set_facecolor('lightgreen')

# Add annotations for peak bin
max_count = max([p.get_height() for p in ax.patches])  # Find max height
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

# Customize title and labels
plt.title(" Age Distribution of Candidates ", fontsize=14, fontweight='bold', color="darkblue")
plt.xlabel("Age", fontsize=12, fontweight='bold', color="darkgreen")
plt.ylabel("Count", fontsize=12, fontweight='bold', color="darkred")

# Add grid and adjust layout
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Subtle gridlines
plt.xticks(fontsize=10, fontweight="bold", color="black")  # Custom X-axis ticks
plt.yticks(fontsize=10, fontweight="bold", color="black")  # Custom Y-axis ticks

# Show the plot
plt.show()
```

## <font color="darkblue"><span style="font-family:Courier;">**📌 Phase-wise Candidate Distribution**</font></span>

### <font color="darkblue"><span style="font-family:Courier;">**🔍 Steps in Analysis:**</font></span>    
- **Count Distribution** – The number of candidates in each phase is calculated using `value_counts()` and sorted in order.  
- **Data Display** – The phase-wise candidate count is printed for reference.  
- **Visualization** – A **bar chart** is created using `Seaborn` to represent the **candidate distribution across phases**.  
- **Customization** – Labels, gridlines, and color palettes are adjusted for better readability.  

### <font color="darkblue"><span style="font-family:Courier;">**📊 Insights:**</font></span>

## <font color="darkblue"><span style="font-family:Courier;">**📌 Seats Won by Each Party**</font></span>

- **Bar Chart (Horizontal)**: Helps in better readability, especially when comparing multiple categories.  
- **Dynamic Color Palette**: Uses Seaborn's "pastel" color scheme for enhanced visual appeal.

  - **Chart title ("Seats Won by Each Party")** styled in purple for emphasis.

# <font color="darkblue"><span style="font-family:Courier;">***📌 State-wise Winning Party Trends***</font></span>

This analysis focuses on **state-wise trends** by visualizing the number of seats won by the top 5 political parties (excluding Independent candidates) in the **2024 Lok Sabha elections**.  

### <font color="darkblue"><span style="font-family:Courier;">***🔹 Steps Performed:***</font></span>  
1. **Excludes Independent Candidates**  
   - Filters out candidates labeled as **"Independent"** from the dataset.  

2. **Identifies the Top 5 Winning Parties**  
   - Counts the total seats won by each party.  
   - Selects the **top 5 parties** based on the highest number of seats won.  

3. **Analyzes State-wise Party Wins**  
   - Groups data by **State and Party** to count the number of seats won per party in each state.  
   - Creates a **pivot table (unstacked)** to prepare data for visualization.  

4. **Visualizes the Data Using a Stacked Bar Chart**  
   - Uses a **stacked bar plot** to represent state-wise party dominance.  
   - Applies the **"tab10" colormap** for clear differentiation.  

5. **Enhances Readability**  
   - Customizes **labels, titles, and gridlines** for better clarity.  
   - Rotates **state labels (90°)** to prevent overlap.  
   - Positions the **legend outside the plot** for better visibility.  

### <font color="darkblue"><span style="font-family:Courier;">***📊 Purpose***</font></span> 
This visualization helps in understanding the **regional strongholds** of major political parties and how the **seat distribution varies across states** in the 2024 elections.  


```python
# Exclude Independent candidates
df = df_results[df_results['Party'] != "Independent"]
# Count total wins per party
party_wins = df['Party'].value_counts()
# Select the top 5 winning parties (excluding Independents)
top5_parties = party_wins.nlargest(5).index
# Filter the dataframe to include only the top 5 parties
df_filtered = df[df['Party'].isin(top5_parties)]
# Count the number of wins for each of the top 5 parties per state
state_party_wins = df_filtered.groupby(['State', 'Party']).size().unstack(fill_value=0)
# Plot the stacked bar chart
ax=state_party_wins.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='magma')
ax.set_facecolor('lightgrey')
# Customize the plot
plt.xlabel("State", fontsize=12, fontweight="bold", color="magenta")
plt.ylabel("Number of Seats Won", fontsize=12, fontweight="bold", color="magenta")
plt.title("State-wise Winning Party Trends",fontsize=16, fontweight="bold", color="brown")
plt.xticks(rotation=90)

plt.legend(title="Winning Party", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle="--", alpha=0.7)
# Show the plot
plt.tight_layout()
plt.show()
```

# <font color="darkblue"><span style="font-family:Courier;">***📌 📊 Party-wise Victory Rate***</font></span>  

This analysis calculates and visualizes the **victory rate** of political parties .  

### <font color="darkblue"><span style="font-family:Courier;">***📌 Insight***</font></span>  
This plot helps in understanding which parties had the **highest success rate** in converting nominations into victories.  


```python
# Count total candidates fielded by each party
total_candidates = df_candidates['Party'].value_counts()

# Count total seats won by each party
total_wins = df_winners['Winning Party'].value_counts()

# Calculate victory rate (Success Rate)
victory_rate = (total_wins / total_candidates) * 100

# Remove NaN values (Parties that fielded candidates but won no seats)
victory_rate = victory_rate.dropna()

# Sort by victory rate (descending order)
victory_rate = victory_rate.sort_values(ascending=False)
colors = sns.color_palette("pastel", len(victory_rate))
# Plot the bar chart
plt.figure(figsize=(12, 6),facecolor='lightgray')
ax=victory_rate.plot(kind='bar', color=colors)
ax.set_facecolor('lightgreen')
# Customize the plot
plt.xlabel("Political Party", fontsize=12, fontweight="bold", color="pink")
plt.ylabel("Victory Rate (%)", fontsize=12, fontweight="bold", color="pink")
plt.title("Party-wise Victory Rate", fontsize=16, fontweight="bold", color="red")
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

# <font color="darkblue"><span style="font-family:Courier;">***📌 📊 Election Results Analysis: Age, Gender & Regional Trends***</font></span>

This block analyzes election results based on three key factors: **age, gender, and regional dominance** of winning candidates.

---

### <font color="darkblue"><span style="font-family:Courier;">***📌 📊Insight***</font></span>
Identifies which **age group** has the most successful candidates.

---

Reveals the **gender representation** in election victories.

---

Highlights **which political parties dominate in different regions** of India.

---

## <font color="darkblue"><span style="font-family:Courier;">***📌 🎯 Key Takeaways***</font></span>
✅ Helps understand **age trends** in winning candidates.  
✅ Shows **gender-wise electoral success**.  
✅ Provides insights into **regional party performance**.  

This analysis can be useful for **political strategy, election studies, and voter behavior analysis**. 🚀


```python


# Merge winners with candidates data to get age
df_winners = df_winners.merge(df_candidates[['State', 'Constituency_No', 'Constituency', 'Candidate Name', 'Age', 'Gender']], 
                              left_on=['State', 'PC No', 'PC Name', 'Winning Candidate'], 
                              right_on=['State', 'Constituency_No', 'Constituency', 'Candidate Name'],
                              how="left",suffixes=('', '_candidate'))

# Drop unnecessary columns
df_winners.drop(columns=['Constituency_No', 'Constituency', 'Candidate Name'], inplace=True)


### ✅ 1. Candidate Success by Age
# Define age groups
bins = [25, 35, 45, 55, 65, 75, 100]
labels = ['25-34', '35-44', '45-54', '55-64', '65-74', '75+']
df_winners['Age Group'] = pd.cut(df_winners['Age'], bins=bins, labels=labels)

# Count winners in each age group
age_group_counts = df_winners['Age Group'].value_counts().sort_index()

# Plot age group success trends
plt.figure(figsize=(10, 5),facecolor='lightgray')
ax=sns.barplot(x=age_group_counts.index, y=age_group_counts.values,hue=age_group_counts.index,  # Assign hue
            palette="coolwarm")
ax.set_facecolor('lightgreen')
plt.xlabel("Age Group", fontsize=12, fontweight="bold", color="grey")
plt.ylabel("Number of Winners", fontsize=12, fontweight="bold", color="grey")
plt.title("Candidate Success by Age Group", fontsize=16, fontweight="bold", color="red")
plt.grid(axis='y', linestyle="--", alpha=0.7)


plt.show()

### ✅ 2. Gender-wise Winning Trends
gender_counts = df_winners['Gender'].value_counts()

# Plot Gender-wise Winning Trends
plt.figure(figsize=(6, 6),facecolor='lightgray')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['yellow', 'pink'], startangle=90)
plt.title("Gender-wise Winning Trends", fontsize=16, fontweight="bold", color="purple")
plt.show()


### ✅ 3. Regional Dominance – North/South/East/West Party Performance
# Define states into regions
north_states = ["Punjab", "Haryana", "Uttar Pradesh", "Rajasthan", "Delhi", "Himachal Pradesh", "Jammu & Kashmir", "Uttarakhand"]
south_states = ["Tamil Nadu", "Kerala", "Andhra Pradesh", "Karnataka", "Telangana"]
east_states = ["West Bengal", "Odisha", "Jharkhand", "Bihar", "Assam"]
west_states = ["Maharashtra", "Gujarat", "Goa", "Madhya Pradesh", "Chhattisgarh", "Rajasthan"]

# Create a region column
def assign_region(state):
    if state in north_states:
        return "North"
    elif state in south_states:
        return "South"
    elif state in east_states:
        return "East"
    elif state in west_states:
        return "West"
    else:
        return "Other"

df_winners['Region'] = df_winners['State'].apply(assign_region)

# Count winning parties in each region
regional_party_counts = df_winners.groupby(["Region", "Winning Party"]).size().unstack().fillna(0)

# Plot Regional Dominance
regional_party_counts.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set2')
plt.xlabel("Region", fontsize=12, fontweight="bold", color="red")
plt.ylabel("Number of Seats Won", fontsize=12, fontweight="bold", color="red")
plt.title("Regional Dominance of Parties (North/South/East/West)",fontsize=16, fontweight="bold", color="brown")
plt.legend(title="Winning Party", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

# <font color="darkblue"><span style="font-family:Courier;">***📌 🗳️ Analysis of Party Performance***</font></span>

This analysis examines the performance of political parties in the **Lok Sabha Election 2024**, comparing the number of candidates they fielded with the seats they won. The insights help evaluate the **success rate** of each party.

---

- This metric indicates how efficient each party was in converting contested seats into victories.  

---

### <font color="darkblue"><span style="font-family:Courier;">***📌 📊Insight***</font></span>

- A higher number of contested candidates does not always translate into more wins.  
- Some parties have a **high success rate**, meaning they strategically contested and secured victories efficiently.  

---

### <font color="darkblue"><span style="font-family:Courier;">***📌 🔍 Conclusion***</font></span> 
This analysis provides a **quantitative overview** of how political parties performed in **Lok Sabha Election 2024**. The success rate metric helps assess which parties were more effective in winning seats relative to the number of candidates they fielded.


```python
# Load Excel files
file1 = "candidates_with_phase.csv"
file2 = "results_2024.csv"
file3 = "results_2024_winners.csv"

df_candidates = pd.read_csv(file1)
df_results = pd.read_csv(file2)
df_winners = pd.read_csv(file3)
# ✅ 1. Count Total Candidates Fielded by Each Party
party_candidates = df_candidates.groupby("Party")["Candidate Name"].count().reset_index()
party_candidates.rename(columns={"Candidate Name": "Total Candidates"}, inplace=True)

# ✅ 2. Count Total Seats Won by Each Party
party_wins = df_winners.groupby("Winning Party")["State"].count().reset_index()
party_wins.rename(columns={"State": "Seats Won", "Winning Party": "Party"}, inplace=True)

# ✅ 3. Merge Both DataFrames to Compare
party_performance = pd.merge(party_candidates, party_wins, on="Party", how="left").fillna(0)
party_performance["Seats Won"] = party_performance["Seats Won"].astype(int)

# ✅ 4. Calculate Success Rate (% of Candidates Who Won)
party_performance["Success Rate (%)"] = (party_performance["Seats Won"] / party_performance["Total Candidates"]) * 100

# ✅ 5. Sort by Seats Won (Top 10)
party_performance = party_performance.sort_values(by="Seats Won", ascending=False).head(10)

# ✅ 6. Plot Candidates Contested vs. Seats Won
fig, ax1 = plt.subplots(figsize=(12, 6),facecolor='lightgray')

bar1=sns.barplot(x="Party", y="Total Candidates", data=party_performance, color="lightgreen", label="Total Candidates", ax=ax1)
bar2=sns.barplot(x="Party", y="Seats Won", data=party_performance, color="orange", label="Seats Won", ax=ax1)
ax1.set_facecolor('lightyellow')
ax1.set_xlabel("Party", fontsize=12, fontweight="bold", color="brown")
ax1.set_ylabel("Count", fontsize=12, fontweight="bold", color="brown")
ax1.set_title("Candidates Contested vs. Seats Won (Top 10 Parties)", fontsize=16, fontweight="bold", color="purple")
ax1.legend()


plt.xticks(rotation=45)
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

# <font color="darkblue"><span style="font-family:Courier;">***📌 Analysis of Closest Contest***</font></span> 


## <font color="darkblue"><span style="font-family:Courier;">***🔍 Key Insights***</font></span>
- The closest contests often occur in **highly competitive seats**.  
- Even a **few hundred votes** can decide a parliamentary seat!  
- This analysis helps in understanding **swing constituencies** where minor shifts can impact election outcomes.  



```python


# ✅ 1. Select Relevant Columns
df_winners_filtered = df_winners[["State", "PC Name", "Winning Candidate", "Winning Party", 
                                   "Runner-up Canddiate", "Runner-up Party", "Margin Votes"]]

# ✅ 2. Convert Margin Votes to Numeric (If Not Already)
df_winners_filtered = df_winners_filtered.copy()
df_winners_filtered["Margin Votes"] = pd.to_numeric(df_winners_filtered["Margin Votes"], errors="coerce")

# ✅ 3. Get the Top 10 Closest Contests (Smallest Margins)
df_closest_contests = df_winners_filtered.nsmallest(10, "Margin Votes")

# ✅ 4. Plot Smallest Winning Margins
plt.figure(figsize=(12, 6),facecolor='lightgray')
ax=sns.barplot(data=df_closest_contests, y="PC Name", x="Margin Votes", hue="Winning Party", dodge=False, palette="twilight")
ax.set_facecolor('lightgreen')
plt.xlabel("Winning Margin (Votes)", fontsize=12, fontweight="bold", color="purple")
plt.ylabel("Constituency", fontsize=12, fontweight="bold", color="purple")
plt.title("Top 10 Closest Contests – Smallest Winning Margins", fontsize=16, fontweight="bold", color="pink")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.legend(title="Winning Party")
plt.show()
```

# <font color="darkblue"><span style="font-family:Courier;">***📊 Analysis of Landslide Wins***</font></span>
